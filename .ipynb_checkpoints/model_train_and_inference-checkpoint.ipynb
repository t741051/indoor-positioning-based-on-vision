```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale=1./255,  # 对图像进行缩放，将像素值归一化到[0,1]之间
    rotation_range=10,  # 随机旋转角度范围
    width_shift_range=0.2,  # 随机水平平移的范围
    height_shift_range=0.2,  # 随机竖直平移的范围
    shear_range=0.1,  # 剪切强度
    zoom_range=0.2,  # 随机缩放范围
    horizontal_flip=False,  # 随机水平翻转
    fill_mode='nearest'  # 填充方式
)

# test_datagen = ImageDataGenerator(rescale=1./255)
train_dir = 'Dataset/train/'

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical'
)


# 创建验证集的 ImageDataGenerator 对象
validation_datagen =  ImageDataGenerator(rescale=1./255)

val_dir = 'Dataset/val/'

# 加载验证集
validation_generator = validation_datagen.flow_from_directory(
    val_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical'
)
```

```python
from tensorflow.keras.applications.inception_resnet_v2 import InceptionResNetV2
from tensorflow.keras.models import Model

# 建立 InceptionResNetV2 模型，載入 ImageNet 預訓練權重
input_shape = (128, 128, 3)
num_classes = 17
base_model = InceptionResNetV2(input_shape=input_shape, include_top=False, weights='imagenet')

# 添加自定義的輸出層
x = base_model.output
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.Dense(1024, activation='relu')(x)
predictions = tf.keras.layers.Dense(num_classes, activation='softmax')(x)

# 構建完整模型
model = Model(inputs=base_model.input, outputs=predictions)

# 鎖定 InceptionResNetV2 的卷積層權重，只訓練自定義的輸出層
for layer in base_model.layers:
    layer.trainable = False

checkpoint = ModelCheckpoint(filepath='model.h5', monitor='val_accuracy', save_best_only=True)   
    
# 編譯模型
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# 訓練模型
# history = model.fit(x_train, y_train_one_hot, epochs=30, validation_split = 0.3, callbacks=[checkpoint])
history = model.fit(train_generator, epochs=15, validation_data=validation_generator, callbacks=checkpoint)
```

```python
val_acc = max(history.history['val_accuracy'])
train_acc = history.history['accuracy'][history.history['val_accuracy'].index(val_acc)]
print("最高驗證準確度：", val_acc)
print("對應的訓練準確度：", train_acc)
```

```python
from tensorflow.keras.models import load_model
model = load_model('model.h5')
```

```python
# 載入測試集資料
test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory(
        'Dataset/test',
        target_size=(128, 128),
        batch_size=32,
        class_mode='categorical',
        shuffle=False)

# 預測測試集圖片的類別
preds = model.predict(test_generator)

# 將預測結果轉換為類別標籤
predicted_classes = np.argmax(preds, axis=1)

# 取得類別名稱列表
class_names = list(test_generator.class_indices.keys())

# 顯示預測結果
for i, img_path in enumerate(test_generator.filepaths):
    print("預測:", class_names[predicted_classes[i]], " - 圖片:", img_path)
```

```python
loss, acc = model.evaluate(test_generator)
print('Test accuracy:', acc)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 載入測試資料集
test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory(
    'Dataset/test',
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical')

# 取得預測結果
predictions = model.predict(test_generator)
predicted_labels = np.argmax(predictions, axis=1)

# 取得實際標籤
true_labels = test_generator.classes

# 取得標籤對應的類別名稱
label_map = {v: k for k, v in test_generator.class_indices.items()}
predicted_class_names = [label_map[label] for label in predicted_labels]
true_class_names = [label_map[label] for label in true_labels]

# 繪製混淆矩陣
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(true_class_names, predicted_class_names)
plt.imshow(cm, cmap=plt.cm.Blues)
plt.colorbar()
tick_marks = np.arange(len(test_generator.class_indices))
plt.xticks(tick_marks, test_generator.class_indices.keys(), rotation=90)
plt.yticks(tick_marks, test_generator.class_indices.keys())
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# 繪製一些預測結果的圖片
fig, axes = plt.subplots(4, 4, figsize=(12, 12))
axes = axes.ravel()

for i in range(16):
    # 隨機選一張圖片
    idx = np.random.randint(len(test_generator.filenames))
    img_path = test_generator.filepaths[idx]
    img = plt.imread(img_path)
    
    # 繪製預測結果和實際標籤的比較圖
    axes[i].imshow(img)
    axes[i].set_title(f"True: {true_class_names[idx]}, Predicted: {predicted_class_names[idx]}")
    axes[i].axis('off')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix
from tensorflow.keras.utils import to_categorical

# 載入測試資料集
test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory(
    'Dataset/test',
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical')

# 取得預測結果和信心度
predictions = model.predict(test_generator)
predicted_labels = np.argmax(predictions, axis=1)
predicted_probas = np.max(to_categorical(predicted_labels, num_classes=num_classes) * predictions, axis=1)

# 取得實際標籤
true_labels = test_generator.classes

# 取得標籤對應的類別名稱
label_map = {v: k for k, v in test_generator.class_indices.items()}
predicted_class_names = [label_map[label] for label in predicted_labels]
true_class_names = [label_map[label] for label in true_labels]

# 繪製混淆矩陣
cm = confusion_matrix(true_class_names, predicted_class_names)
plt.imshow(cm, cmap=plt.cm.Blues)
plt.colorbar()
tick_marks = np.arange(len(test_generator.class_indices))
plt.xticks(tick_marks, test_generator.class_indices.keys(), rotation=90)
plt.yticks(tick_marks, test_generator.class_indices.keys())
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# 繪製一些預測結果的圖片和信心度
fig, axes = plt.subplots(4, 4, figsize=(12, 12))
axes = axes.ravel()

for i in range(16):
    # 隨機選一張圖片
    idx = np.random.randint(len(test_generator.filenames))
    img_path = test_generator.filepaths[idx]
    img = plt.imread(img_path)
    
    # 取得預測結果和信心度
    predicted_class_name = predicted_class_names[idx]
    true_class_name = true_class_names[idx]
    predicted_proba = predicted_probas[idx]
    
    # 繪製預測結果和實際標籤的比較圖以及信心度
    axes[i].imshow(img)
    axes[i].set_title(f"True: {true_class_name}, Predicted: {predicted_class_name}\nConfidence: {predicted_proba:.2f}")
    axes[i].axis('off')

plt.show()
```
