```python
import os
import numpy as np
from PIL import Image

# 設定圖片目錄路徑
image_dir = 'Dataset/train'

# 建立空的資料集
data = []
labels = []

# 迭代處理每個子目錄
for subdir in os.listdir(image_dir):
    subdir_path = os.path.join(image_dir, subdir)

    # 檢查是否為目錄
    if not os.path.isdir(subdir_path):
        continue

    # 設定 label 為資料夾名稱
    label = subdir

    # 迭代處理每個子目錄中的所有圖片檔案
    for file in os.listdir(subdir_path):
        file_path = os.path.join(subdir_path, file)

        # 檢查檔案是否為圖片
        if not file_path.endswith('.jpg') and not file_path.endswith('.jpeg') and not file_path.endswith('.png'):
            continue

        # 讀取圖片檔案
        try:
            img = Image.open(file_path)
        except Exception as e:
            print(f"Error reading {file_path}: {e}")
            continue

        # 轉換圖片資料為 NumPy 陣列
        img_data = np.array(img)

        # 將圖片資料加入資料集
        data.append(img_data)

        # 將 label 加入 labels
        labels.append(label)
```

```python
# 將資料集與 labels 轉換成 NumPy 陣列
data = np.array(data)
labels = np.array(labels)

# 儲存資料集與 labels 到 .npy 檔案中
if os.path.exists('train_image.npy'):
    # 刪除已經存在的文件
    os.remove('train_image.npy')
if os.path.exists('train_label.npy'):
    # 刪除已經存在的文件
    os.remove('train_label.npy')
np.save('train_image.npy', data)
np.save('train_label.npy', labels)
```

```python
import numpy as np

# 讀取 .npy 檔案
data = np.load('train_image.npy')
labels = np.load('train_label.npy')

# 檢查讀取後的形狀
print('test_image shape:', data.shape)
print('test_label shape:', labels.shape)

# 列出前 10 筆圖片及其對應標籤
for i in range(210,220):
    img = Image.fromarray(data[i])
    label = labels[i]
    display(img)
    print('Label:', label)
```
